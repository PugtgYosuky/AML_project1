### 2.4 - Correlated features

```python
corr = data.corr()
plt.figure(figsize=(20, 15))
plt.title('Correlation between the different numerical features')
sns.heatmap(data=corr.round(2), annot=True)
```

```python
corr
```

```python
correlated_features = corr.abs() > 0.8
plt.figure()
sns.heatmap(data=corr[correlated_features].round(2), annot=True)
plt.title('Correlation greater than 0.8')
plt.grid(True)
```

```python
for col in correlated_features.columns:
    if correlated_features[col].sum() > 1:
        print(f'{col}')
    cols = correlated_features.index[correlated_features[col]]
    for f in cols:
        if f != col:
            print(f'  -{f}')
```
